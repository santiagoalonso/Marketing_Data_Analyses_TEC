## Synthetic data 

Synthetic controls, in brief, is weighting some variables in control units to make them as close as possible to the treatment unit.

Mixtape, Scott Cunnigham

Python implementation: https://matheusfacure.github.io/python-causality-handbook/15-Synthetic-Control.html

```python
cigar = pd.read_csv("smoking.csv")
cigar
```

```python
plt.figure(figsize=(7,4))
cigar_r = (
  cigar
 .assign(california = np.where(cigar["california"], "California", "Other States"))
 .groupby(["year", "california"])
 ["cigsale"]
 .mean()
 .reset_index()
)
idx = cigar_r.loc[:,"california"] == "California"
plt.plot(cigar_r.loc[idx,"year"], cigar_r.loc[idx,"cigsale"], label = "California")
plt.plot(cigar_r.loc[~idx,"year"], cigar_r.loc[~idx,"cigsale"], label = "Other States")
plt.vlines(x=1988, ymin=40, ymax=140, linestyle=":", lw=2, label="Proposition 99")
plt.ylabel("Cigarette Sales Trend")
plt.title("Gap in per-capita cigarette sales (in packs)")
plt.legend();  
```

```python
#Data setup
features = ["cigsale", "retprice"] #Feature to make the units as comparable before the intervention
inverted = (cigar.query("~after_treatment") # filter pre-intervention period
            .pivot(index='state', columns="year")[features] # make one column per year and one row per state
            .T) # flip the table to have one column per state (the column name "3" is california)
y = inverted[3].values # state of california
X = inverted.drop(columns=3).values  # other states
```

```python
#Weights to make the synthetic control via linear regression
weights_lr = LinearRegression(fit_intercept=False).fit(X, y).coef_
weights_lr.round(3) #Each weight is a state
```

```python
#Now multiply each control unit cigsale (not california) by the weights (with dot product)
calif_synth_lr = (cigar.query("~california")
                  .pivot(index='year', columns="state")["cigsale"]
                  .values.dot(weights_lr))
```

```python
#Compare with the previous one
plt.figure(figsize=(7,4))
plt.plot(cigar.query("california")["year"], cigar.query("california")["cigsale"], label="California")
plt.plot(cigar.query("california")["year"], calif_synth_lr, label="Synthetic Control")
plt.vlines(x=1988, ymin=40, ymax=140, linestyle=":", lw=2, label="Proposition 99")
plt.ylabel("Gap in per-capita cigarette sales (in packs)")
plt.legend();
```
